## Data Exploration

```python
#Import Libraries
import pandas as pd
import seaborn as sns
```

```python
# Read data
data = pd.read_csv("../Final-Project/WA_Fn-UseC_-HR-Employee-Attrition.csv")
data.head()
```

```python
# Display all column data
pd.set_option('display.max_columns', 999)
```

```python
# Review for all data
data.head()
```

```python
#Shape of data
data.shape
```

```python
# Describe dataset
data.describe()
```

```python
# Describe dataset
data.describe()
```

```python
# Data counts - attrition
data['Attrition'].value_counts()
```

```python
#Visualize attrition count 
sns.countplot(data['Attrition'])
```

```python
#Show attrition and retention of employees by age
import matplotlib.pyplot as plt
fig_dims = (12, 4)
fig, ax = plt.subplots(figsize=fig_dims)
sns.countplot(x='Age', hue='Attrition', data = data, ax = ax,  edgecolor=sns.color_palette("dark", n_colors = 1));
```

```python
#Print all of the object data types and their unique values
for column in data.columns:
    if data[column].dtype == object:
        print(str(column) + ' : ' + str(data[column].unique()))
        print(data[column].value_counts())
        print("_________________________________________________________________")
```

```python
#Get the correlation of the columns
data.corr()
```

```python
#Visualize the correlation
plt.figure(figsize=(14,14))  #14in by 14in
sns.heatmap(data.corr(), annot=True, fmt='.0%')
```
